```python
# LEGAL
import numpy as np
import matplotlib.pyplot as plt
 
# Create dataset
height = [0.0674, 0.0791,0.0718,0.0855,0.1037,0.1263 ]
bars = ('RetinaNet TrueColor', 'RetinaNet Agriculture','Augmented True Color','Augmented Agriculture','RetinaNet TrueColor Multi','RetinaNet Agriculture Multi')
x_pos = np.arange(len(bars))
 
# Create bars
plt.figure(figsize=(9,6)) 
plt.title('Dice Coefficient Comparison Between all RetinaNet Models for Legal Mines Detection')
plt.bar(x_pos, height)

plt.ylim(0, 0.15)
# Create names on the x-axis
plt.xticks(x_pos, bars)
plt.xticks(rotation=45)
# Show graphic
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
 
# Read CSV into pandas
data = {'Model': ['RetinaNet TrueColor', 'RetinaNet Agriculture','Augmented True Color','Augmented Agriculture','RetinaNet TrueColor Multi','RetinaNet Agriculture Multi'],
        'Dice': [0.0674, 0.0791,0.0718,0.0855,0.1037,0.1263]}
# Create DataFrame.
df = pd.DataFrame(data)

 
model = df['Model'].head(6)
dice = df['Dice'].head(6)
 
# Figure Size
fig, ax = plt.subplots(figsize =(5, 4))
 
# Horizontal Bar Plot
ax.barh(model, dice,color =["#c90808","#c90808","#c90808","#c90808","#730303","#730303"])
 
# Remove axes splines
#for s in ['top', 'bottom', 'left', 'right']:
#    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
#ax.xaxis.set_ticks_position('none')
#ax.yaxis.set_ticks_position('none')
ax.set_xlim(right=0.13)
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
#ax.grid(b = True, color ='grey',
#        linestyle ='-.', linewidth = 0.5,
#        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()-0.03, i.get_y()+0.5,
             str((i.get_width())),
             fontsize = 10, fontweight ='bold',
             color ='white')
 
# Add Plot Title
#ax.set_title('Dice Coefficient Comparison Between all RetinaNet Models for Legal Mines Detection',
#             loc ='left', )
 
# Add Text watermark
fig.text(0.8, -0.01, 'Dice Coefficient', fontsize = 12,
         color ='black', ha ='right', va ='bottom',
        alpha = 0.7)


plt.tight_layout()
plt.savefig("DiceLegalRetinaMulti.jpg")
# Show Plot
plt.show()
```
