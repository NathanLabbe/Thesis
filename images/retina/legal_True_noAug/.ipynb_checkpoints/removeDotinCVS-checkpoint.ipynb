```python
def clean(input):
    tmpFile = "tmp.csv"
    independant = []
    with open(input, "r") as file, open(tmpFile, "w",newline='') as outFile:
        reader = csv.reader(file, delimiter=',')
        writer = csv.writer(outFile, delimiter=',')
        header = next(reader)
        writer.writerow(header)
        for row in reader:
            x=0
            colValues = []
            t=0
            for col in row:
                if x==3 or x==4 or x==5 or x==6:
                    if col=='':
                        t=1
                        colValues.append(col)
                    else:
                        colValues.append(int(float(col)))
                else:
                    if (x==1) or (x==2):
                        print("skip")
                    else:
                        colValues.append(col)
                if x==0:
                    if col not in independant:
                        independant.append(col)
                x+=1
            if t==0:
                if (colValues[1]<colValues[3]) and (colValues[2]<colValues[4]):
                    writer.writerow(colValues)
    print(len(independant))
    os.rename(tmpFile,'test.csv')
```

```python
clean('val_annotations.csv')
```
