```python
from PIL import Image
import os
import os.path
import random
import shutil
```

```python
for element in os.listdir('./AGRICULTURE/illegal'):
    if os.path.isdir(element):
        print("'%s' un dossier" % element)
    else:
        im = Image.open('./AGRICULTURE/illegal/'+element)
        element2 = element.replace('.jpg','.tif')
        print(element2)
        im.save('./unet/AllOriginals/'+element2)  # or 'test.tif'
```

```python
for element in os.listdir('./AGRI_ILE_MASK'):
    if os.path.isdir(element):
        print("'%s' un dossier" % element)
    else:
        im = Image.open('./AGRI_ILE_MASK/'+element)
        element2 = element.replace('.png','.tif')
        print(element2)
        im.save('./unet/AllMasks/'+element2)  # or 'test.tif'
```

```python
for element in os.listdir('./unet/AllOriginals/'):
    if not os.path.isfile('./unet/AllMasks/'+element):
        print (element)
        os.remove("./unet/AllOriginals/"+element)
```

```python
nbImages = 0
for element in os.listdir('./unet/tmpMasks/'):
    nbImages+=1
print(nbImages)

trainSize = 80
trainSize = int(nbImages*trainSize/100)

for x in range(trainSize):
    file = random.choice(os.listdir("./unet/tmpMasks")) #change dir name to whatever
    shutil.move("./unet/tmpMasks/"+file, "./unet/train/"+file)
    
for element in os.listdir('./unet/tmpMasks/'):
    shutil.move("./unet/tmpMasks/"+element, "./unet/test/"+element)
    
    
```

```python
for element in os.listdir('./unet/train/'):
    if not os.path.isfile('./unet/train/mask_'+element):
        shutil.copy2("./unet/AllOriginals/"+element,'./unet/train/mask_'+element)
```

```python
for element in os.listdir('./unet/test/'):
    if not os.path.isfile('./unet/test/mask_'+element):
        shutil.copy2("./unet/AllOriginals/"+element,'./unet/test/mask_'+element)
```
